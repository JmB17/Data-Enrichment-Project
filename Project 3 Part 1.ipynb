```python
import pandas as pd
import numpy as np
import matplotlib as mpl
```

```python
akas = pd.read_csv('Data/title.akas.tsv',sep='\t')
basics = pd.read_csv('Data/title.basics.tsv',sep='\t')
ratings = pd.read_csv('Data/title.ratings.tsv',sep='\t')
```

## Akas Data

```python
akas.head()
```

```python
akas.info()
```

Drop all rows without region set to 'US'

```python
filter_akas = akas[akas['region'] == 'US']
filter_akas
```

```python
akas = filter_akas
```

```python
akas['region'].value_counts()
```

Replace /N with np.nan

```python
akas = akas.replace({'\\N':np.nan})
```

```python
akas.head()
```

```python
akas.info()
```

## Basics Data

```python
basics.head()
```

```python
basics.info()
```

```python
# Keep only movies from US
basics_filter = basics['tconst'].isin(akas['titleId'])
basics_filter
```

```python
basics = basics[basics_filter]
basics.head()
```

```python
# Replace "\N" with np.nan
basics = basics.replace({'\\N':np.nan})
```

```python
basics.head()
```

```python
# Dropping rows with null values in runtimeMinutes and genres 
basics = basics.dropna(subset = ["runtimeMinutes","genres"])
```

```python
print(basics['runtimeMinutes'].isna().sum())
print(basics['genres'].isna().sum())
```

```python
# Keeping rows with titleType == Movie
basics_movie = basics[basics['titleType'] == 'movie']
basics_movie
```

```python
basics = basics_movie
```

```python
print(basics['titleType'].value_counts())
print(basics.shape)
```

```python
# Convert startYear column to float data type
basics['startYear'] = basics['startYear'].astype(float)
```

```python
basics.head()
```

```python
basics.info()
```

```python
basics['startYear'].dtype
```

```python
# Filter the dataframe using startYear. Keep years between 2000-2021 (Including 2000 and 2021)
basics_year = basics.loc[basics["startYear"].between(2000, 2021)]
basics_year
```

```python
basics = basics_year
```

```python
basics.head()
```

```python
basics['startYear'].value_counts()
```

```python
basics.info()
```

```python
basics['genres'].value_counts()
```

```python
basics.drop(basics.loc[basics['genres']== 'Documentary'].index, inplace=True)
```

```python
basics['genres'].value_counts()
```

```python
basics.info()
```
