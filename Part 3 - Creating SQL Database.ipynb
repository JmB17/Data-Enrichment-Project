# Normalize Generes column

Convert the single string of genres from title basics into 2 new tables.

- title_genres: with the columns:
     - tconst
     - genre_id

- genres:
     - genre_id
     - genre_name


```python
tmdb_data['genres_split'] = tmdb_data['genres'].str.split(', ')
tmdb_data
```

```python
exploded_genres = tmdb_data.explode('genres_split')
exploded_genres
```

```python
title_genres = exploded_genres[['imdb_id','genres_split']].copy()
title_genres
```

```python
# Make a list of all characters to replace
to_replace = ['[{','}','}]','{',']']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    title_genres['genres_split'] = title_genres['genres_split'].str.replace(char,'',regex=False)
    
title_genres['genres_split'].head()
```
